```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
df['Species']=df['Species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
df.head()
```

```python
X=df.drop('Species',axis=1)
y=df.Species
```

```python
print(y),print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=17)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
X_test
```

```python
X_train
```

```python
y_train
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(X_test)
```

```python
X_test
```

```python
pedicted_output=clf.predict(X_test)
pedicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pedicted_output)
print(cm)
```

```python
## plotting the confution matrix
plt.figure(figsize=(5,5))
sns.heatmap(cm,annot=True)
plt.title('Confution matrix')
plt.xlabel('Predicted value')
plt.ylabel('Actual value or true value')
```
